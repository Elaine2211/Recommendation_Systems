## Content Based Recommendation System
---

The object of this system is: If a user gives us a movie title, we can recommend movies sharing similar plot summaries based on the their overviews.

```python
print("Overview:",movies_cleaned.iloc[11,:]['overview'])
print("Title:",movies_cleaned.iloc[11,:]['original_title']) 
```

```python
# Use NLP to vectorize words into matrix
from sklearn.feature_extraction.text import TfidfVectorizer


tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Fill NAN with empty string
movies_cleaned['overview']=movies_cleaned['overview'].fillna('')
```

```python
tfv_matrix=tfv.fit_transform(movies_cleaned['overview'])
```

```python
tfv_matrix
```

```python
tfv_matrix.toarray()
```

```python
tfv_matrix.shape
```

```python
# Using sklearn.metrics.pairwise submodule implements utilities to evaluate pairwise distances or affinity of sets of samples
from sklearn.metrics.pairwise import sigmoid_kernel
sig=sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
sig[0]
```

```python
indices=pd.Series(movies_cleaned.index,index=movies_cleaned['original_title']).drop_duplicates()
```

```python
# Define a function for recommendation
def give_rec(title,sig=sig):
    #Get the index corresponding to the title
    idx=indices[title]
    
    #Get the pairwise similarity scores
    sig_scores=list(enumerate(sig[idx]))
    
    #Sort the movies, using sorted for reordering a list or sequence, using sort_values for reordering DF
    sig_scores=sorted(sig_scores,key=lambda x:x[1],reverse=True)
 
    #Scores of the 10 most similar movies
    sig_scores_most=sig_scores[1:11]
    
    #Movie indices
    movie_indices=[i[0] for i in sig_scores_most]
    
    #Top 10 most similar movies
    return movies_cleaned['original_title'].iloc[movie_indices]
```

```python
#Test the system using Titanic
give_rec('Titanic')
```
