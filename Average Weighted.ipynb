```python
import numpy as np
import pandas as pd
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
print('Shape of credits:',credits.shape)
print('Shape of movies:',movies.shape)
```

```python
#Combine the two dataframes
movies_merge=pd.merge(left=movies, how='inner', right=credits, left_on='id',right_on='movie_id').drop(['movie_id'],axis=1)
```

```python
movies_cleaned=movies_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
```

```python
movies_cleaned.head()
```

```python
movies_cleaned.info()
```

```python
# Use Calculation to get the Weighted Average
v=movies_cleaned['vote_count']
r=movies_cleaned['vote_average']
c=movies_cleaned['vote_average'].mean()
m=movies_cleaned['vote_count'].quantile(0.7)  
# m is the minimum votes required to be listed in the top 250
```

```python
movies_cleaned['weighted_average']=((r*v)+(c*m))/(v+m)
```

```python
movies_cleaned.head()
```

```python
movies_ranking=movies_cleaned.sort_values('weighted_average',ascending=False)
movies_ranking[['original_title','vote_count','vote_average','weighted_average','release_date','popularity']].head(20)
```

```python
movies_ranking_pop=movies_cleaned.sort_values('popularity',ascending=False)
movies_ranking_pop[['original_title','vote_count','vote_average','weighted_average','release_date','popularity']].head(20)
```

```python
# Take into account both weighted_average and popularity (5-5)
# Standerdize the two columns since they are in different ranges

from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler()
movies_scaled=sc.fit_transform(movies_cleaned[['weighted_average','popularity']])
movies_normalized=pd.DataFrame(movies_scaled,columns=['weighted_average','popularity'])
movies_normalized.head()
```

```python
movies_cleaned[['normalized_weighted_average','normalized_popularity']]=movies_normalized
movies_cleaned.head()
```

```python
movies_cleaned['score']=movies_cleaned['normalized_weighted_average']*0.5+movies_cleaned['normalized_popularity']*0.5
movies_ranking_score=movies_cleaned.sort_values(['score'],ascending=False)
movies_ranking_score=movies_ranking_score[['original_title','vote_count','vote_average','weighted_average','release_date','popularity','score']].head(20)
```
